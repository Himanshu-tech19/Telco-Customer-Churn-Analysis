```python
avg_charges = df.groupby('tenure_range')['monthlycharges'].mean()
avg_charges
```

```python
plt.figure(figsize=(7,5))
bars = plt.bar(
    avg_charges.index,
    avg_charges.values
)
plt.xlabel("Tenure Group")
plt.ylabel("Average Monthly Charges")
plt.title("Average Monthly Charges by Tenure Group")
```

```python
import plotly.express as px
fig = px.bar(
    avg_charges.reset_index(),
    x='tenure_range',
    y='monthlycharges',
    title="Average Monthly Charges by Tenure Group",
    text='monthlycharges'
)
fig.show()
```

```python
tenure_labels = ['0-12 months', '13-36 months', '37+ months']


avg_charges = pd.Series(avg_charges, index=tenure_labels)

fig, ax = plt.subplots()

bars = ax.bar(tenure_labels, avg_charges)

# value labels
for bar in bars:
    y = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width()/2,
        y,
        f"{y:.2f}",
        ha='center',
        va='bottom'
    )

# annotation
max_value = avg_charges.max()
max_index = avg_charges.idxmax()
x_pos = tenure_labels.index(max_index)

ax.annotate(
    "Highest Average\nMonthly Charges",
    xy=(x_pos, max_value),
    xytext=(x_pos, max_value + 10),
    arrowprops=dict(arrowstyle="->"),
    ha='center'
)

plt.ylabel("Average Monthly Charges")
plt.title("Average Monthly Charges by Tenure Group")
plt.show()
```

**Tasks 5: Advanced Analysis**

```python
churn_rate = df.groupby('tenure_range')['churn'].apply(
    lambda x:(x == 'Yes').mean() * 100)

churn_rate
```

```python
churn_by_gender = (
    df.groupby('gender')['churn']
      .apply(lambda x: (x == 'Yes').mean() * 100)
)

churn_by_gender
```

```python
churn_by_senior = (
    df.groupby('seniorcitizen')['churn']
      .apply(lambda x: (x == 'Yes').mean() * 100)
)

churn_by_senior
```

```python
churn_by_payment = (
    df.groupby('paymentmethod')['churn']
      .apply(lambda x: (x == 'Yes').mean() * 100)
      .reset_index(name='churn_rate_%')
)

churn_by_payment
```

```python
churn_by_contract = (
    df.groupby('contract', observed=False)['churn']
      .apply(lambda x: (x == 'Yes').mean() * 100)
      .reset_index(name='churn_rate_%')
)

churn_by_contract
```

```python
early_risk = df[
    (df['churn'] == 'No') &                      # Not churned yet
    (df['tenure_range'] == '0-12 months') &      # Early stage
    (df['contract'] == 'Month-to-month') &      # Low commitment
    (df['paymentmethod'] == 'Electronic check')  # High-churn payment type
]

early_risk
```

```python
mid_churn = df[
    (df['tenure_range'] == '13-36 months') &
    (df['churn'] == 'Yes')
]
```

```python
mid_churn['contract'].value_counts(normalize=True) * 100
```

```python
mid_churn['paymentmethod'].value_counts(normalize=True) * 100
```

```python
mid_churn['monthlycharges'].describe()
```

```python
mid_non_churn = df[
    (df['tenure_range'] == '13-36 months') &
    (df['churn'] == 'No')
]

mid_non_churn['monthlycharges'].describe()
```

```python
mid_churn['seniorcitizen'].value_counts(normalize=True) * 100
```

```python
mid_risk = df[
    (df['churn'] == 'No') &
    (df['tenure_range'] == '13-36 months') &
    (df['paymentmethod'] == 'Electronic check') &
    (df['contract'] == 'Month-to-month') &
    (
        (df['onlinesecurity'] == 'No') |
        (df['techsupport'] == 'No')
    )
]
mid_risk
```

```python
late_churn = df[
    (df['tenure_range'] == '37+ months') &
    (df['churn'] == 'Yes')
]
```

```python
late_churn['contract'].value_counts(normalize=True) * 100
```

```python
late_churn['paymentmethod'].value_counts(normalize=True) * 100
```

```python
late_churn['monthlycharges'].describe()
```

```python
late_non_churn = df[
    (df['tenure_range'] == '13-36 months') &
    (df['churn'] == 'No')
]

late_non_churn['monthlycharges'].describe()
```

```python
late_churn['seniorcitizen'].value_counts(normalize=True) * 100
```

```python
late_risk = df[
    (df['churn'] == 'No') &
    (df['tenure_range'] == '37+ months') &
    (df['monthlycharges'] > df['monthlycharges'].quantile(0.75)) &
    (df['onlinesecurity'] == 'No') &
    (df['techsupport'] == 'No')
]

late_risk
```

```python
print("Early phase risk:", early_risk.shape[0])
print("Mid phase risk:", mid_risk.shape[0])
print("Late phase risk:", late_risk.shape[0])
```

```python
at_risk_customers=pd.concat([early_risk,mid_risk,late_risk])
at_risk_customers
```

```python
at_risk_customers= at_risk_customers.drop_duplicates()

```

```python
at_risk_customers.to_csv(
    'at_risk_customers_final.csv',
    index=False
)
```
